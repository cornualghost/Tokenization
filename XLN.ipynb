```python
import pandas as pd
from transformers import XLNetTokenizer
from sklearn.pipeline import Pipeline
from sklearn.base import TransformerMixin

# Carica il tokenizzatore pre-addestrato di XLNet
tokenizer = XLNetTokenizer.from_pretrained('xlnet-base-cased')

# Definizione di un Transformer personalizzato per utilizzare il tokenizzatore XLNet
class XLNetTokenization(TransformerMixin):
    def transform(self, X, **transform_params):
        # Applica la tokenizzazione XLNet a ogni elemento
        return [tokenizer.encode(text, add_special_tokens=True) for text in X]

    def fit(self, X, y=None, **fit_params):
        return self

# Creazione della pipeline
pipeline = Pipeline([
    ('xlnet_tokenization', XLNetTokenization())
])

# Supponiamo di avere un DataFrame pandas con una colonna di testo
data = {'text': ["Esempio di testo per tokenizzare.", "Un altro testo da tokenizzare."]}
df = pd.DataFrame(data)

# Applicazione della pipeline
df['tokenized'] = pipeline.fit_transform(df['text'])

# Visualizza il DataFrame
print(df)
```
